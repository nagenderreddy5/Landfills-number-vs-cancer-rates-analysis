```python
# Setup 

import pandas as pd
import numpy as np
```

```python
# Storing Path

cancer_data = "Resources/County_Cancer_Data.csv"
```

```python
# Reading in the data

county_cancer_df = pd.read_csv(cancer_data)
county_cancer_df.head()
```

```python
# Remove unnecessary columns from the dataframe

reduced_county_cancer_df = county_cancer_df[["State","Age-Adjusted Incidence Rate(†) - cases per 100,000","Average Annual Count","Recent Trend", ]]
reduced_county_cancer_df.head()
```

```python
def del_superscript (x):
    return x.split("(")[0]

def to_str (x):
    return f"{x}" 

def state (x):
    return x.split(",")[1]
```

```python
reduced_county_cancer_df["State"] = reduced_county_cancer_df["State"].map(to_str)

reduced_county_cancer_df["State"] = reduced_county_cancer_df["State"].copy().map(del_superscript)

# reduced_county_cancer_df.dtypes

reduced_county_cancer_df.head()
```

```python
# Split out county and stae into separate columns

state_breakout = reduced_county_cancer_df["State"].copy().str.split(",", n = 1, expand = True)

state_breakout = state_breakout.rename(columns = {0:"County",1:"State"})

# reduced_county_cancer_df["State"] = reduced_county_cancer_df["State"].map(state)

# reduced_county_cancer_df["County"] = state_breakout[0]

# reduced_county_cancer_df["State"] = state_breakout[1]

# reduced_county_cancer_df.drop(columns =["State"], inplace = True)

state_breakout.head()
```

```python

# reduced_county_cancer_df = county_cancer_df[["State","Age-Adjusted Incidence Rate(†) - cases per 100,000","Average Annual Count","Recent Trend", ]]

reduced_county_cancer_df = reduced_county_cancer_df.rename(columns = {'State': 'State and County','Age-Adjusted Incidence Rate(†) - cases per 100,000':'Age-Adjusted Incidence Rate - cases per 100,000'} )

reduced_county_cancer_df.head()
```

```python
# Combine data frames 

combined_df = pd.merge(state_breakout,reduced_county_cancer_df, left_index = True, right_index = True)

# state_breakout.join(reduced_county_cancer_df, how = 'outer') 
combined_df.head()
```

```python
combined_df.groupby(["State", "Recent Trend"])["County"].count()
```

```python
del combined_df['State and County']

combined_df.head()
```

```python
# Rid of Kansas and Minnesota data, as they are excluded from this report

combined_df.drop(combined_df.loc[combined_df['State'] ==' Kansas'].index, inplace = True)

combined_df.drop(combined_df.loc[combined_df['State'] == ' Minnesota'].index, inplace = True)


# Rid of "3 or fewer" rows uner "Average Annual Count" column
combined_df.drop(combined_df.loc[combined_df['Average Annual Count'] == '3 or fewer'].index, inplace = True)


# print (combined_kansas_drop)
# print (combined_df)
```
